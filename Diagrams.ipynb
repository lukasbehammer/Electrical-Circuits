```python
import schemdraw
import schemdraw.elements as elm

elm.style(elm.STYLE_IEC)
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
with schemdraw.Drawing(file='Circuit_1.svg') as d:
    d += (A := elm.MeterA().right(4))
    d += (R := elm.Resistor().down().label(r'$R = 1 \/ \mathrm{M立}$', loc='bottom', rotate=90, ofst=0.2))
    d += elm.Line().left(4)
    d += (Q := elm.BatteryCell().up().reverse())

    d += elm.CurrentLabelInline(direction='in').at(A).label(r'$I$')
    d += elm.CurrentLabel(reverse=True, ofst=0.7).at(Q).label(r'$U_q = 9 \/ \mathrm{V}$', rotate=90)
    d += elm.CurrentLabel(reverse=True, top=False, ofst=0.5).at(R).label(r'$U_R$', rotate=90)
```

```python
with schemdraw.Drawing(file='Circuit_2.svg') as d:
    d += (L1 := elm.Line().right())
    d += elm.Line()
    d += elm.MeterV().down()
    d += (L2 := elm.Line().left())
    d += elm.Line()
    d += (Q := elm.BatteryCell().up().reverse())
    d += (R := elm.Resistor().endpoints(L1.end, L2.end).label(r'$R = 1 \/ \mathrm{M立}$', loc='bottom', rotate=90, ofst=0.2))

    d += elm.CurrentLabelInline(direction='in', ofst=0.2).at(L1).label(r'$I$')
    d += elm.CurrentLabel(reverse=True, ofst=0.7).at(Q).label(r'$U_q = 9 \/ \mathrm{V}$', rotate=90)
    d += elm.CurrentLabel(reverse=True, top=False, ofst=0.5).at(R).label(r'$U_R$', rotate=90)
```

```python
with schemdraw.Drawing(file='Circuit_3.svg') as d:
    d += (Q := elm.SourceV().up())
    d += (L := elm.Line().right(4))
    d += (R1 := elm.Resistor().down().label(r'$R_1 = 330 \/ \mathrm{立}$', loc='bottom', rotate=90, ofst=0.2))
    d += (R2 := elm.Resistor().length(4).to(Q.start).label(r'$R_2 = 680 \/ \mathrm{立}$', loc='bottom', ofst=0.2))

    d += elm.CurrentLabelInline(direction='in').at(L).label(r'$I$')
    d += elm.CurrentLabel(reverse=True, ofst=0.7).at(Q).label(r'$U_q = 10 \/ \mathrm{V}$', rotate=90)
    d += elm.CurrentLabel(reverse=True, top=False, ofst=0.5).at(R1).label(r'$U_1$', rotate=90)
    d += elm.CurrentLabel(reverse=True, top=False, ofst=0.5).at(R2).label(r'$U_2$')
```

```python
with schemdraw.Drawing(file='Circuit_4.svg') as d:
    d += (Q := elm.SourceV().up(4))
    d += (L1 := elm.Line().right())
    d += elm.Line()
    d += (R2 := elm.Resistor().down(4).label(r'$R_2 = 680 \/ \mathrm{立}$', loc='bottom', rotate=90, ofst=0.2))
    d += (L2 := elm.Line().left())
    d += elm.Line()
    d += (R1 := elm.Resistor().endpoints(L1.end, L2.end).label(r'$R_1 = 330 \/ \mathrm{立}$', loc='bottom', rotate=90, ofst=0.2))

    d += elm.CurrentLabelInline(direction='in', ofst=0.2).at(L1).label(r'$I$')
    d += elm.CurrentLabelInline(direction='in', ofst=1.2).at(R1).label(r'$I_1$')
    d += elm.CurrentLabelInline(direction='in', ofst=1.2).at(R2).label(r'$I_2$')
    d += elm.CurrentLabel(reverse=True, ofst=0.7).at(Q).label(r'$U_q = 10 \/ \mathrm{V}$', rotate=90)
    d += elm.CurrentLabel(reverse=True, top=False, ofst=0.5).at(R1).label(r'$U_1$', rotate=90)
    d += elm.CurrentLabel(reverse=True, top=False, ofst=0.5).at(R2).label(r'$U_2$', rotate=90)
```

```python
with schemdraw.Drawing(file='Circuit_5.svg') as d:
    d += (Q := elm.BatteryCell().up().reverse())
    d += (L1 := elm.Line().right())
    d += (R1 := elm.Resistor().down().label(r'$R_1 = 100 \/ \mathrm{k立}$', loc='bottom', rotate=90, ofst=0.2))
    d += (LED1 := elm.LED())
    d += elm.Line().right()
    d += (bjt := elm.BjtNpn(circle=True))
    d += elm.LED().up().toy(LED1.start).reverse().flip()
    d += elm.Resistor().toy(R1.start).label(r'$R_2 = 1 \/ \mathrm{k立}$', loc='bottom', rotate=90, ofst=0.2)
    d += (L2 := elm.Line().to(R1.start))
    d += elm.Line().at(bjt.emitter).tox(Q.start)
    d += elm.Line().to(Q.start)

    d += elm.CurrentLabelInline(direction='in', ofst=0.2).at(L1).label(r'$I$')
    d += elm.CurrentLabel(reverse=True, ofst=0.7).at(Q).label(r'$U_q = 9 \/ \mathrm{V}$', rotate=90)
```

```python
with schemdraw.Drawing(file='Circuit_6.svg') as d:
    d += (Q := elm.SourceV().up())
    d += (R := elm.Resistor().right(5).label(r'$R = 1 \/ \mathrm{k立}$', ofst=0.2))
    d += (C := elm.Capacitor().down().label(r'$C = 1 \/ \mathrm{袖F}$', rotate=90, ofst=0.2))
    d += (L1 := elm.Line().length(5).to(Q.start))
    d += (L2 := elm.Line().at(C.start).right(1))
    d += (D1 := elm.Dot(open=True))
    d += (L3 := elm.Line().at(C.end).right(1))
    d += (D2 := elm.Dot(open=True))

    d += elm.CurrentLabelInline(direction='in', ofst=1).at(R).label(r'$I$')
    d += elm.CurrentLabel(reverse=True, ofst=0.7).at(Q).label(r'$U_{pp} = 2 \/ \mathrm{V}$', rotate=90)
    d += elm.CurrentLabel(reverse=True, top=False, ofst=0.5).at(R).label(r'$U_R$')
    d += elm.CurrentLabel(ofst=1).at(C).label(r'$U_C$')
```
